# Correlation matrix for numerical features-
Correlation analysis helps identify relationships between numerical features.
A correlation matrix and heatmap can reveal if certain features, like RAM (GB) and Price (Euro), or Inches and Weight (kg), have strong associations.

```python
#Correlation matrix for numerical features
correlation_matrix = data[['CPU_Frequency (GHz)', 'RAM (GB)', 'Weight (kg)', 'Price (Euro)']].corr()

# Plot a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Specs and Price')
plt.show()

```

the average price per company

```python
# Group by company to see average price per brand
average_price_by_company = data.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

# Display the result
print(average_price_by_company)

# Plotting the average price per company
plt.figure(figsize=(12, 6))
average_price_by_company.plot(kind='bar', color='skyblue')
plt.title('Average Laptop Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (Euro)')
plt.xticks(rotation=45)
plt.show()

```

 Analyze Price by CPU Company and Type


```python
 # Analyze Price by CPU Company and Type
# Group by CPU_Company to see average price per CPU brand
average_price_by_cpu_company = data.groupby('CPU_Company')['Price (Euro)'].mean().sort_values(ascending=False)

# Display the result
print(average_price_by_cpu_company)

# Group by CPU_Type to see average price per CPU type
average_price_by_cpu_type = data.groupby('CPU_Type')['Price (Euro)'].mean().sort_values(ascending=False)

# Display the result
print(average_price_by_cpu_type)

# Plotting the average price per CPU company
plt.figure(figsize=(12, 6))
average_price_by_cpu_company.plot(kind='bar', color='lightblue')
plt.title('Average Laptop Price by CPU Company')
plt.xlabel('CPU Company')
plt.ylabel('Average Price (Euro)')
plt.xticks(rotation=45)
plt.show()

# Plotting the average price per CPU type
plt.figure(figsize=(12, 6))
average_price_by_cpu_type.plot(kind='bar', color='orange')
plt.title('Average Laptop Price by CPU Type')
plt.xlabel('CPU Type')
plt.ylabel('Average Price (Euro)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouping by 'Product' to find the top products based on price
# Load data from CSV
data_sales_latest = pd.read_csv('laptop_prices.csv')

top_products_by_price = data_sales_latest.groupby('Product')['Price (Euro)'].mean().sort_values(ascending=False)

# Convert to DataFrame for better readability
top_products_df = top_products_by_price.reset_index()

# Display the top 5 most expensive laptops
print("Top 5 Most Expensive Laptops:")
print(top_products_df.head())

# You can also visualize the top 10 most expensive laptops
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top_products_df['Product'].head(10), top_products_df['Price (Euro)'].head(10), color='skyblue')
plt.xlabel('Average Price (Euro)')
plt.ylabel('Laptop Model')
plt.title('Top 10 Most Expensive Laptops')
plt.gca().invert_yaxis()  # Invert y-axis to show highest price at the top
plt.show()
```
